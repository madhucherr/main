importing libraies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

importing the dataset

```python
df=pd.read_csv('Salary_Data.csv')
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(x)
print(y)
```

plotting linear graph

```python
plt.plot(x,y)
plt.title('linear graph')
plt.show()
```

plotting Scatter graph

```python
plt.scatter(x,y)
plt.title('Scatter plot')
plt.show()
```

plotting histogram

```python
dd = df.head(3)
plt.hist(dd)
plt.title('histogram')
plt.show()
```

plotting pie chart

```python
plt.pie(x)
plt.title('pie chart')
plt.show()
```

plotting bar graph

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```
